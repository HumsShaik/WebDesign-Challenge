```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities_df = pd.read_csv("Resources/cities.csv")
cities_df
```

```python
len(cities_df)
```

```python
cities_df.to_html("Resources/cities_data.html")
```

## Plotting the Data

## Latitude vs. Temperature Plot

```python
# Incorporate the other graph properties
# Save the figure
# Show plot
plot_lat = cities_df["Lat"]
plot_temp = cities_df["Max Temp"]

plt.scatter(plot_lat,plot_temp)
plt.title('Max Temperature vs. City Latitude')
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.savefig('Images/plot_latitude_vs_temp')
plt.show()
```

## Latitude vs. Humidity Plot

```python
plot_humidity = cities_df["Humidity"]

plt.scatter(plot_lat,plot_humidity)
plt.title('Humidity vs. City Latitude')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.savefig('Images/plot_latitude_vs_humidity')
plt.show()
```

## Latitude vs. Cloudiness Plot

```python
plot_cloudiness = cities_df["Cloudiness"]

plt.scatter(plot_lat,plot_cloudiness)
plt.title('Cloudiness vs. City Latitude')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.savefig('Images/plot_latitude_vs_cloudiness')
plt.show()
```

## Latitude vs. Wind Speed Plot

```python

wind_speed = cities_df["Wind Speed"]

plt.scatter(plot_lat,wind_speed)
plt.title('Wind Speed vs. City Latitude')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.savefig('Images/plot_latitude_vs_windspeed')
plt.show()
```
